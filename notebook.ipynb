```python
-- TEAM_WITH_MAJORITY_POSSESSION
WITH possesssion_counts AS (
	SELECT 
		TEAM_NAME_HOME AS team,
		COUNT(*) AS game_count
	FROM
		SOCCER.TBL_UEFA_2021
	WHERE
		POSSESSION_HOME > POSSESSION_AWAY
	GROUP BY
		TEAM_NAME_HOME

	UNION ALL

	SELECT
		TEAM_NAME_AWAY AS team,
		COUNT(*) AS game_count
	FROM
		SOCCER.TBL_UEFA_2021
	WHERE
		POSSESSION_AWAY > POSSESSION_HOME
	GROUP BY
		TEAM_NAME_AWAY)
	
SELECT
	team AS TEAM_NAME,
	SUM(game_count) AS Game_Count
FROM
	possesssion_counts
GROUP BY
	team
LIMIT 1
```
